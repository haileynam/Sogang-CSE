```python
def drawData(x_c1, x_c2):
    plt.plot(x_c1[:,1], x_c1[:,2], '+b')
    plt.plot(x_c2[:,1], x_c2[:,2], 'or')
```

```python
# def drawLine(w, col):
#     for x in range (-5, 11, 1):
#         y = w[0][0]/w[0][2]-w[0][1]/w[0][2]*x
#         plt.plot(x, y, col)
def drawLine(w, col):
    x = np.array([int(x) for x in range(-5, 11, 1)])
    y = w[0][0]/w[0][2]-w[0][1]/w[0][2]*x
    plt.plot(x, y, col)
```

```python
def drawDataHyperPlane(x_c1, x_c2, w, col):
    drawData(x_c1, x_c2)
    drawLine(w, col)
#     plt.show()
```

```python
def Perceptron_mis_yx(y, x, w, n_epoch, eta):
    #n_sample = np.size(y, 1)/2
    #print(n_sample)
    x_c1_1 = x[0 :n_sample, :]
    print(x_c1_1)
    x_c2_1 = x[(n_sample): (n_sample *2+1), :]
    print(x_c2_1)
    yx = np.matlib.repmat(y, 1,3)*x
    print('yx:' ,yx)
    
    print('----- Perceptron_mis_yx() started -----\n')
    for e in range(1, n_epoch+1):
            #wyx = np.dot(np.tile(w, (n_sample*2, 1)), yx, 2)
            #wyx = np.dot(np.matlib.repmat(w, n_sample*2,1), yx, 2)
            wyx = np.multiply(np.matlib.repmat(w, n_sample*2, 1), yx).sum(axis=1)
            yx_misclassified = yx[wyx<0,:]
            #print(wyx)
            #print(yx_misclassified)
            w = w + sum(yx_misclassified)
            err = sum(wyx<0)
            print('err =', err)
            plt.xlim([-5, 10])
            plt.ylim([-5, 10])
            drawDataHyperPlane(x_c1_1, x_c2_1, w, 'k')
            if err == 0:
                break
    plt.show()
    return e
```

```python
def Perceptron_l2(y, x, w, n_epoch, eta):
    x_c1_2 = x[0:n_sample,:]
    x_c2_2 = x[n_sample: n_sample*2+1,:]
    
    print('----- Perceptron_l2() started -----\n')
    for e in range(1, n_epoch+1):
        #wx = np.dot(np.matlib.repmat(w, n_sample*2, 1), x)
        wx = np.multiply(np.matlib.repmat(w, n_sample*2, 1), x).sum(axis = 1)
        wx = np.reshape(wx, (40,1))
        #print('wx\n',wx)
        
        diff = (np.matlib.repmat((y-wx), 1, 3))*x
        #print(diff)
        w = w + eta*sum(diff)
        err1 = 0.5*((y-wx)**2).sum()
        err2 = sum(y*wx < 0)
        print('err(', e, ') =', err1, err2)
        plt.xlim([-5, 10])
        plt.ylim([-5, 10])
        drawDataHyperPlane(x_c1_2, x_c2_2, w, 'k')
        if err2 == 0: 
            break
    plt.show()
    return e
```
